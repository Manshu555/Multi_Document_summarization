```python
import pandas as pd
import re
import numpy as np
from transformers import BartTokenizer
from tqdm import tqdm

# Load tokenizer
tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")

# Drop rows with missing article or summary
df = df.dropna(subset=['article', 'highlights']).reset_index(drop=True)

# Remove empty or very short texts (optional but helpful)
df = df[df['article'].str.len() > 100]
df = df[df['highlights'].str.len() > 10]
df = df.reset_index(drop=True)

# Clean function
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)                # Remove extra whitespace, newlines
    text = re.sub(r'[^a-zA-Z0-9.,!?\'\"\s]', '', text)  # Remove unwanted symbols
    text = text.strip()
    return text

# Apply cleaning
df['clean_article'] = df['article'].apply(clean_text)
df['clean_summary'] = df['highlights'].apply(clean_text)
```

```python
newdf = df[['clean_article', 'clean_summary']]
```
